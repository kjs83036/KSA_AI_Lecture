```
import numpy as np
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
x = np.array(x)

y = [1, 5 ,3 ,2 ,3]
y = np.array(y)

plt.plot(x, y, marker='.', ls="")
plt.show()


w = [ -1, -0.5, 0, 0.5, 1]
loss = []
for i in w:
    
    tmp = np.sum((x * i - y)**2) / len(x)
    loss.append(tmp)
print(w)
print(loss)
plt.plot(w, loss)
plt.show()
```

```
import numpy as np
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

x_train = x
y_train = y

model = Sequential()
model.add(Dense(1, input_dim=1))

sgd = SGD(lr=0.05)
model.compile(loss='mse', optimizer=sgd)

# prints summary of the model to the terminal
model.summary()

model.fit(x_train, y_train, epochs=200)

y_predict = model.predict(np.array([5]))
print(y_predict)
```

###선형회귀

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import RMSprop
import numpy as np

x_data = np.array([[73., 80., 75.],
                   [93., 88., 93.],
                   [89., 91., 90.],
                   [96., 98., 100.],
                   [73., 66., 70.]])
y_data = np.array([[152.],
                   [185.],
                   [180.],
                   [196.],
                   [142.]])

model = Sequential()
model.add(Dense(1, input_dim=3))

rmsprop = RMSprop(lr=1e-2)
model.compile(loss='mse', optimizer=rmsprop)
model.fit(x_data, y_data, epochs=1000)

y_predict = model.predict(np.array([[95., 100., 80]]))
print(y_predict)
```

```
'''
opencv-python 라이브러리 설치:

    pip install opencv-python 
'''

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# 테스트 이미지 불러오기
img = cv2.imread('/content/drive/My Drive/lecture/l_d/p5/lena.jpg', cv2.IMREAD_GRAYSCALE)

# 필터(filter) 정의 
vKernel = np.array([[-1, 0, 1],
                    [-1, 0, 1],
                    [-1, 0, 1]])

hKernel = np.array([[-1,-1,-1],
                    [ 0, 0, 0],
                    [ 1, 1, 1]])

# 컨볼루션(convolution) 연산
vertical = cv2.filter2D(img, -1, vKernel)
horizontal = cv2.filter2D(img, -1, hKernel)
both = vertical + horizontal

# 필터 시각화
import matplotlib.pyplot as plt
%matplotlib inline

plt.imshow(vKernel, interpolation='bilinear', cmap='gray')
plt.axis('off')
plt.show()

# 컨볼루션 결과 출력
cv2_imshow(img)
cv2_imshow(vertical)
cv2_imshow(horizontal)
cv2_imshow(both)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
