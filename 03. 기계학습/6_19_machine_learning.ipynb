Confusion Matrix

```python
import pandas as pd

data = {"y_Predicted" : [1, 1, 0, 1, 0, 1, 1, 0, 1, 0, 0, 0],
       "y_Actual" : [1, 0, 0, 1, 0, 1, 0, 0, 1, 0, 1, 0]
       }
df = pd.DataFrame(data,
                  columns=["y_Actual",
                           "y_Predicted"
                          ]
                 )
confusion_matrix = pd.crosstab(df["y_Actual"],
                               df["y_Predicted"],
                               rownames=["Actual"],
                               colnames=["Predicted"])
confusion_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(confusion_matrix, annot=True)
plt.show()
```

```python
confusion_matrix2 = pd.crosstab(df["y_Actual"],
                                df["y_Predicted"],
                                rownames=["Actual"],
                                colnames=["Predicted"],
                                margins=True
                               )
confusion_matrix2
```

```python
from pandas_ml import ConfusionMatrix

ConfusionMatrix3 = ConfusionMatrix(df["y_Actual"],
                                 df["y_Predicted"]
                                 )
ConfusionMatrix3
```

```python
ConfusionMatrix3.print_stats()
```
