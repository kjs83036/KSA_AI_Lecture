```
from konlpy.tag import Okt
from gensim.models import word2vec


model = word2vec.Word2Vec.load(
    '/content/drive/My Drive/lecture/l_d/p6/wiki.model'
    )
def print_emergency(text):
    print(text)
    
    count = 0
    
    node = okt.pos(text, norm=True, stem=True)
    for word, form in node:
        if (form == 'Noun' or form == 'Verb' or
            form == 'Adjectiv' or form == 'Adverb'):
            print('-', word, ':', model.wv.similarity(word, '급하다'))
            if model.wv.similarity(word, '급하다') > 0.5:
                count += 1
    return print('count, node : ', count, len(node))
    
    
        
print_emergency('문제가 생겼어요. 빨리 해결해 주세요.')
print_emergency('사용 방법을 잘 모르겠습니다.')
```
