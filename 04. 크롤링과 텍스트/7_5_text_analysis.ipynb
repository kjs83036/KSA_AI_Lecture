```
import codecs

from bs4 import BeautifulSoup
from konlpy.tag import Okt
from gensim.models import word2vec


fp = codecs.open("/content/drive/My Drive/lecture/l_d/p4/ch02/BEXX0003.txt",
                encoding="utf-16"
                )

soup = BeautifulSoup(fp, "html.parser")
body = soup.select_one("body > text")
text = body.getText()
```

```
okt = Okt()
results=[]
lines = text.split("\r\n")
for line in lines:
    malist = okt.pos(line, norm=True, stem=True)
    r = []
    for word in malist:
        if not word[1] in ["Josa", "Eomi", "Punctuation"]:
            r.append(word[0])
    rl = (" ".join(r).strip())
    results.append(rl)
    print(rl)
```

```
wakati_file = "toji.wakati"
with open(wakati_file, 'w', encoding='utf-8') as fp:
    fp.write('\n'.join(results))
    
data = word2vec.LineSentence(wakati_file)
model = word2vec.Word2Vec(data, size=200, window=10, hs=1, min_count=2, sg=1)
model.save("/content/drive/My Drive/lecture/l_d/p4/ch02/toji.model")
print('ok')
```

##문장의 유사도
###레벤슈타인거리

```
import numpy as np


def calc_distance(a, b):
    if a == b:
        return 0
    a_len = len(a)
    b_len = len(b)
    if a == "":
        return b_len
    if b == "":
        return a_len
    
    matrix = [[] for i in range(a_len + 1)]
    #print("1", matrix)
    for i in range(a_len + 1):
        matrix[i] = [0 for j in range(b_len + 1)]
    #print("2", matrix)
    for i in range(a_len + 1):
        matrix[i][0] = i
    #print("3", matrix)
    for j in range(b_len + 1):
        matrix[0][j] = j
    #print("4", matrix)
    for i in range(1, a_len + 1):
        ac = a[i - 1]
        for j in range(1, b_len + 1):
            bc = b[j - 1]
            cost = 0 if (ac == bc) else 1
            matrix[i][j] = min([
                matrix[i - 1][j] + 1,
                matrix[i][j - 1] + 1,
                matrix[i - 1][j - 1] +cost
            ])
        print("5", np.array(matrix))
    return matrix[a_len][b_len]
print(calc_distance("가나다라", "가마바라"))

samples = ["신촌역", "신천군", "신천역", "신발", "마곡역"]
base = samples[0]
r = sorted(samples, key=lambda n: calc_distance(base, n))
for n in r:
    print(calc_distance(base, n), n)
```

###N-gram 유사도

```
def ngram(s, num):
    res = []
    slen = len(s) - num + 1
    for i in range(slen):
        ss = s[i:i + num]
        res.append(ss)
    return res

def diff_ngram(sa, sb, num):
    a = ngram(sa, num)
    b = ngram(sb, num)
    r = []
    cnt = 0
    for i in a:
        for j in b:
            if i == j:
                cnt +=1
                r.append(i)
    return cnt / len(a), r
```

```
a = "파이썬으로 하는 빅데이터 분석과 머신러닝은 매우 쉽습니다."
b = "빅데이터 분석과 머신러닝은 파이썬을 이용하여 매우 쉽게 할 수 있습니다."

r_2gram, word_2gram = diff_ngram(a, b, 2)
print("2_gram : ", r_2gram, word_2gram)
r_3gram, word_3gram = diff_ngram(a, b, 3)
print("3_gram :", r_3gram, word_3gram)
r_5gram, word_5gram = diff_ngram(a, b, 5)
print("3_gram :", r_5gram, word_5gram)
```
