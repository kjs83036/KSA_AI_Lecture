##BeautifulSoup

```
from bs4 import BeautifulSoup


html="""
<html><body>
<h1>h1</h1>
<p>p</p>
<p>p2</p>
</body></html>
"""

soup = BeautifulSoup(html, "html.parser")
h1 = soup.html.body.h1
p1 = soup.html.body.p
p2 = p1.next_sibling.next_sibling

print(h1)
print(p1)
print(p2)
#print(soup.prettify())
print(soup.body.p.next_sibling.next_sibling)
print(soup.body.p.next_sibling.encode())
```
