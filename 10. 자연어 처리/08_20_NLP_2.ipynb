## 정수 인코딩

```python
text="A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
```

```python
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt')


text = sent_tokenize(text)
print(text)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
nltk.download('stopwords')


vocab = Counter()

sentences = []
stop_words = set(stopwords.words("english"))

for i in text:
    sentence = word_tokenize(i)
    result = []

    for word in sentence:
        word = word.lower()
        if word not in stop_words:
            if len(word) > 2:
                result.append(word)
                vocab[word] = vocab[word] + 1
    sentences.append(result)

print(sentences)
```

```python
print(vocab)
```

```python
vocab_sorted = sorted(vocab.items(), key=lambda x:x[1], reverse=True)
print(vocab_sorted)
```

```python
word_to_index = {}
i = 0
for (word, frequency) in vocab_sorted:
    if frequency > 1:
        i += 1
        word_to_index[word] = i
    
print(word_to_index)
```

```python
from konlpy.tag import Okt
import re
okt = Okt()


token = re.sub("(\.)", "", "정부가 발표하는 물가상승률과 소비자가 느끼는 물가상승률은 다르다.")
token = okt.morphs(token)

word2index = {}
bow = []
for voca in token:
    if voca not in word2index.keys():
        word2index[voca] = len(word2index)
        bow.insert(len(word2index) - 1, 1)
    else:
        index = word2index.get(voca)
        bow[index] = bow[index] + 1

print(word2index)
```

```python
from konlpy.tag import Komoran


tagger = Komoran()
tagger.nouns("아침에 해를 보며 버스를 탔다")
```

```python
import re


def get_nouns(text):
    nouns = tagger.nouns(text)
    nouns = [word for word in nouns if len(word) > 1]
    nouns = [word for word in nouns if not re.match(r"\d+", word)]
    return nouns
```

```python
get_nouns("아침에 해를 보며 버스를 탔다")
```

```python
from sklearn.feature_extraction.text import CountVectorizer


cv = CountVectorizer(max_features=1000, tokenizer=get_nouns)
tdm = cv.fit_transform(df["본문"])
cv.get_feature_names()[:10]
```

```python
import nltk
from konlpy.corpus import kobill


doc_ko = kobill.open("1809894.txt")
```

```python
from konlpy.tag import Okt


okt = Okt()
```

```python
n = 0
result = []

while True:
    line = doc_ko.readline()
    if not line:
        break
    n += 1
    if n % 5000 == 0:
        print(f"{n}번째")
    tokenlist = okt.pos(line, stem=True, norm=True)
    temp = []
    for word in tokenlist:
        if word[1] in ["Noun"]:
            temp.append((word[0]))
    
    if temp:
        result.append(temp)
    
doc_ko.close()
```

```python
len(result)
```

```python
from gensim.models import Word2Vec


model = Word2Vec(result, size=100, window=5, min_count=5, workers=5, sg=0)
```

```python
a = model.wv.most_similar("학위")
a
```

```python
import pandas as pd
import numpy as np

path = "/content/drive/My Drive/lecture/l_d/p8/"

ntrain = 150000

data = pd.read_csv(path + "ratings.txt", sep='\t', quoting=3)
# data = pd.DataFrame(np.random.permutation(data))
# trn, tst = data[:ntrain], data[ntrain:]

# header = 'id document label'.split()
# trn.to_csv(path + 'ratings_train.txt', sep='\t', index=False, header=header)
# tst.to_csv(path + 'ratings_test.txt', sep='\t', index=False, header=header)
```

```python
data.head()
```

```python
from konlpy.tag import Okt


okt = Okt()

result = []
doc_list = data["document"].tolist()
```

```python
len(doc_list)
```

```python
from tqdm import tqdm


for doc in tqdm(doc_list):
    malist = okt.pos(doc, norm=True, stem=True)
    r=[]
    for word in malist:
        if not word[1] in ["Josa", "Eomi", "Punctuation"]:
            r.append(word[0])
            rl = (" ".join(r)).strip()
            result.append(rl)
```

```python
import csv

f = open(path + "ratings.txt",'r',encoding = 'utf-8')
rdr = csv.reader(f, delimiter = '\t')
rdw = list(rdr)

print(rdw[:3])

f.close()
```

```python
from konlpy.tag import Okt


okt = Okt()

result = []
for line in rdw:
    malist = okt.pos(line[1], norm=True, stem=True)
    r = []
    for word in malist:
        if not word[1] in ["Josa","Eomi","Punctuation"]:
            r.append(word[0])
            rl = (" ".join(r)).strip()
            result.append(rl)
```

```python
with open(path + "naverMovie.nlp", "w", encoding="utf-8") as f:
    f.write("\n".join(result))
```
